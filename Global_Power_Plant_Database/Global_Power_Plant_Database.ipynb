```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Read Data from CSV File
url = "https://raw.githubusercontent.com/neha-mishra/Evaluation-Projects/main/Global_Power_Plant_Database/database_IND.csv"
df = pd.read_csv(url)

# Adjust index to start from 1 instead of 0
df.index += 1
```

```python
# Remove column with unwanted (String) data like - gppd_idnr, latitude, longitude, owner, source, url, geolocation_source, wepp_id, generation_gwh_2019
# Keep only - 
selected_column = [
    'country',
    'country_long',
    'name',
    'capacity_mw',
    'primary_fuel',
    'other_fuel1',
    'other_fuel2',
    'other_fuel3',
    'commissioning_year',
    'year_of_capacity_data',
    'generation_gwh_2013',
    'generation_gwh_2014',
    'generation_gwh_2015',
    'generation_gwh_2016',
    'generation_gwh_2017',
    'generation_gwh_2018',
    'estimated_generation_gwh'
]
```

```python
global_power_plant_df = df[selected_column].copy()
```

```python
global_power_plant_df.head()
```

```python
global_power_plant_df.describe()
```

# Data Visualization

```python
main_primary_fuel = global_power_plant_df.primary_fuel.value_counts() * 100 / global_power_plant_df.primary_fuel.count()
main_primary_fuel
```

```python
sns.barplot(x = main_primary_fuel, y = main_primary_fuel.index)
plt.title('Main primary fuel')
plt.xlabel('Count (Percentages)');
plt.ylabel('Different type of power plant depands on the type of fuel uses');
```

```python
# Type of power plant and their capacity
sns.scatterplot(x = global_power_plant_df.capacity_mw, y = global_power_plant_df.primary_fuel, s = 150)
plt.title('Type of power plant and capacity');
```

```python
# Different type of primary fuel based power plant and their year of going to the first operation
sns.scatterplot(x = global_power_plant_df.commissioning_year, y = global_power_plant_df.primary_fuel, s = 150);
plt.title('Primary fuel and commission year');
```

```python
# Powerplant capacity
total_power_plant = global_power_plant_df.country_long.value_counts()
total_power_plant
```

```python
total_capacity = global_power_plant_df.capacity_mw.sum()
print('India has total capacity of {} megawatt.'.format(total_capacity))
```
